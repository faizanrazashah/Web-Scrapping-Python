```python
with requests.session() as s:
   # r1 = s.post(login_url, data = payload)
    # print(r1.status_code)
    r2 = s.get(attendance_url)
    # print(r2.status_code)
    soup = BeautifulSoup(r2.content, "html.parser")
    
soup    
```

```python
table = soup.find_all("Strong")
```

```python
cols = [header.text for header in table[0].find_all("td")[:-1]]
cols
```

```python
row1 = [row1.text.replace("\n", "").strip() for row1 in table[0].find_all("td")[0:7]]
row1
```

```python
row2 = [row2.text.replace("\n", "").strip() for row2 in  table[0].find_all("td")[8:15]]
row2
```

```python
row3 = [row3.text.replace("\n", "").strip() for row3 in  table[0].find_all("td")[16:23]]
row3
```

```python
all_rows = [row1, row2, row3]
pprint(all_rows)
```
